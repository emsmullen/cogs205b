```python
#test case 1: negative value when constructing
signal_detection_1 = SignalDetection(hits=10, misses=20, false_alarms=-100, correct_rejections=900)
```

```python
#test case 2: non-integer value when constructing
signal_detection_1 = SignalDetection(hits=10.1, misses=20, false_alarms=100, correct_rejections=900)
```

```python
#construct two proper signal detection objects for the following test cases
signal_detection_1 = SignalDetection(hits=10, misses=20, false_alarms=100, correct_rejections=900)
signal_detection_2 = SignalDetection(hits=50, misses=50, false_alarms=20, correct_rejections=80)
```

```python
#test case 3: check that we overloaded the + operator correctly
signal_detection_3 = signal_detection_1 + signal_detection_2
print(signal_detection_3.hits) # should be 60
print(signal_detection_3.misses) # should be 70
print(signal_detection_3.false_alarms) # should be 120
print(signal_detection_3.correct_rejections) # should be 980
```

```python
#test case 4: check that we overloaded the * operator correctly
signal_detection_4 = signal_detection_1 * 2
print(signal_detection_4.hits) # should be 20
print(signal_detection_4.misses) # should be 40
print(signal_detection_4.false_alarms) # should be 200
print(signal_detection_4.correct_rejections) # should be 1800
```

```python
#test case 5: check that we overloaded the - operator correctly
signal_detection_5 = signal_detection_1 - signal_detection_2
print(signal_detection_5.hits) # should be -40 and thus raise ValueError
print(signal_detection_5.misses) # should be -30 and thus raise ValueError
print(signal_detection_5.false_alarms) # should be 80
print(signal_detection_5.correct_rejections) # should be 820
```
